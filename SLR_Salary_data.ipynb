```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# importing our dataset
df=pd.read_csv("Salary_Data.csv")
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# # Plotting the scatter plot
plt.figure(figsize=(12,4))
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.scatter(df["YearsExperience"],df["Salary"],color="red",marker="+")
```

```python
# Creating a LinearRegression model
reg_model=LinearRegression()

# Fit the model using the scatter plot data
reg_model.fit(df[["YearsExperience"]],df["Salary"])
```

```python
# Make predictions for new data points
# For example, predict income for the 14 year of experience
reg_model.predict([[14]])
```

```python
# Now we will check whether our model is working properly or not, by Putting the Values in Equation y=mx+b 
# Value of m
reg_model.coef_
```

```python
# Value of b 
reg_model.intercept_
```

```python
# by putting these values in equation y=mx+b
m=9449.96232146
x=14
b=25792.20019866871
y=m*x+b
print(y)
```

```python
# Plotting the regression line

plt.figure(figsize=(12,4))
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.scatter(df["YearsExperience"],df["Salary"],color="red",marker="+",label="Actual Data")
plt.plot(df["YearsExperience"],reg_model.predict(df[["YearsExperience"]]),color="blue",label="Regression Line")
plt.legend()
plt.show()
```
